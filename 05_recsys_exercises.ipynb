# Create the rec sys

### LigthFM main attributes
***

* **no_components** (int, optional) – the dimensionality of the feature latent 
embeddings. It is simliar to the concept of `component` in `PCA`. It determines the number of `latent_features` in which we *summarise* the rataing matrix

* **learning_schedule** (string, optional) – one of (`adagrad`, `adadelta`). It is the name of the descent gradient algorithm we use to make the recommendation system converge
* **loss** (string, optional) – one of (`logistic`, `bpr`, `warp`, `warp-kos`): the loss function.
    * **logistic**: useful when both positive (1) and negative (-1) interactions are present.
    * **BPR**: Bayesian Personalised Ranking [1]_ pairwise loss. Maximises the
prediction difference between a positive example and a randomly 
chosen negative example. Useful when only positive interactions 
are present and optimising ROC AUC is desired.
    * **WARP**: Weighted Approximate-Rank Pairwise [2]_ loss. 
Maximises the rank of positive examples by repeatedly sampling negative 
examples until rank violating one is found. Useful when only 
positive interactions are present and optimising the top of 
the recommendation list (precision@k) is desired.
    * **k-OS WARP**: k-th order statistic loss [3]_.
A modification of WARP that uses the k-th positive example for any given user as a basis for pairwise updates.




* **item_alpha** (float, optional) – L2 penalty on item features. Tip: setting this number too high can slow down training. One good way to check is if the final weights in the embeddings turned out to be mostly zero. The same idea applies to the user_alpha parameter.
* **user_alpha** (float, optional) – L2 penalty on user features.
max_sampled (int, optional) 
* **random_state** (int seed, RandomState instance, or None) – The seed of the pseudo random number generator to use when shuffling the data and initializing the parameters.

```
%%time
from lightfm import LightFM
rec_sys = LightFM(loss='warp',learning_schedule='adadelta',random_state=101)
# fit the model
rec_sys.fit(movies['train'], epochs=30, num_threads=2,)
```

## Let's wrap everything into a function

```
def sample_recommendation(model, data, user_ids,top=3):
    # number of users and movies in training data
    n_users, n_items  = data['train'].shape

    
    # generate recommendations for each user we input
    for user_id in user_ids:
        # Movies they already like 
        # Mengambil item_label (judul movie) positif sesuai dengan user_id yg di inputkan
        already_seen = data['item_labels'][data['train'].tocsr()[user_id].indices]
        
        # Movies our model predict they will like
        scores = model.predict(user_id, np.arange(n_items))

        # rank them in order of most liked to least
        top_movies = data['item_labels'][np.argsort(-scores)]
        

        # print out what user has seen an the recommendataions        
        print("User %s"%user_id)
        print("   Already seen:")
        for x in already_seen[:top]:
            print("      %s" %x)
        
        print("Recomended : ")
        for x in top_movies[:top]:
            print("      %s" %x)
```

```
sample_recommendation(rec_sys,movies,[11, 14, 25, 32,50, 80],5)
```
